```python
# Select the relevant columns for the waterfall chart
columns_for_waterfall_chart = [
    'Value:Count_ChildDeliveryEvent_InPublicInstitution',
    'Value:Count_ChildDeliveryEvent_InAnInstitution',
    'Value:Count_PregnancyEvent',
    'Value:Count_ChildDeliveryEvent_AtHome',
    'Value:Count_ChildDeliveryEvent_AtHome_WithStandbyAssist',
    'Value:Count_ChildDeliveryEvent',
    'Value:Count_PregnantWomen_RegisteredForAntenatalCare',
    'Value:Count_DeliveryEvent_Safe_AsFractionOf_Count_DeliveryEvent',
    'Value:Count_PregnantWomen_RegisteredForAntenatalCareWithinFirstTrimester'
]

# Calculate the cumulative sum for the waterfall chart
cumulative_sum = df[columns_for_waterfall_chart].sum().cumsum()

# Create the waterfall chart
plt.figure(figsize=(25, 20))  # Set the size of the chart (width, height)
plt.bar(columns_for_waterfall_chart, df[columns_for_waterfall_chart].sum(), color='green', label='Positive')
plt.bar(columns_for_waterfall_chart, -df[columns_for_waterfall_chart].sum(), color='r', label='Negative')
plt.plot(columns_for_waterfall_chart, cumulative_sum, marker='o', color='black', label='Cumulative')

# Customize the chart
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Waterfall Chart')
plt.legend()#(loc='upper left', bbox_to_anchor=(1, 1))  # Move the legend outside the plot

# Rotate x-axis labels if needed
plt.xticks(rotation=90)

# Show the plot
plt.tight_layout()  # Adjust layout to prevent label cutoff
plt.show()
```

# Pie Chart of Pregnancy
<a id='Pie-chart-of-pregnancy'></a>

```python
# Calculate the total pregnancy value for each category
category_totals = df.groupby('placeName')['Value:Count_PregnancyEvent'].sum()
# Identify categories with a total pregnancy value less than 1,00,000
small_categories = category_totals[category_totals < 1000000]
small_category_names = small_categories.index
df['placeName'] = df['placeName'].apply(lambda x: 'Other' if x in small_category_names else x)
merged_category_totals = df.groupby('placeName')['Value:Count_PregnancyEvent'].sum()
categories = merged_category_totals.index
values = merged_category_totals.values
plt.figure(figsize=(10, 10))
plt.pie(values, labels=categories, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Pregnancy Events by State')
plt.axis('equal')
plt.show()
```
